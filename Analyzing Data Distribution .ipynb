## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here

sns.boxplot(x=df["ConvertedComp"])
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here

# The interquartile range is the difference between the 75th percentile and the 25th percentile.

q1 = df['ConvertedComp'].quantile(0.25)
q3 = df['ConvertedComp'].quantile(0.75)
iqr = q3 - q1
print('iqr value is: ',iqr)
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_limit = q1 - 1.5 * iqr
upper_limit = q3 + 1.5 * iqr

print('lower bounds: ' + str(lower_limit))
print('upper bounds: ' + str(upper_limit))
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
cols = ['ConvertedComp']
((df[cols] < (q1 - 1.5 * iqr)) |(df[cols] > (q3 + 1.5 * iqr))).sum()
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
cols = ['ConvertedComp'] # one or more


df_new = df[~((df[cols] < (q1 - 1.5 * iqr)) |(df[cols] > (q3 + 1.5 * iqr))).any(axis=1)]
df_new.head()
```

```python
#median after
df_new["ConvertedComp"].median()
```

```python
#mean after
df_new["ConvertedComp"].mean()
```
